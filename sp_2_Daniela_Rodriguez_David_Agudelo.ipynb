3. Los ingresos se simulan con una distribución normal:

   - Desviación estándar: `2`  
   - Tamaño de muestra: `30`

### Importación de librerías

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
import pandas as pd
import statsmodels.api as sm
```

### Generación de ingresos simulados con semilla: 911848

```python
semilla = 911848
np.random.seed(semilla)

media_teorica = (semilla % 10) + 2
desviacion_estandar = 2
tamano_muestra = 30

simulacion_ingresos = np.random.normal(loc=media_teorica, scale=desviacion_estandar, size=tamano_muestra)

print(simulacion_ingresos)
```

### Cálculo de medidas de tendencia central

```python
media = np.mean(simulacion_ingresos)
mediana = np.median(simulacion_ingresos)
moda = stats.mode(simulacion_ingresos, keepdims=True)[0][0]
```

### Gráfica de resultados

```python
plt.figure(figsize=(10, 5))
sns.histplot(simulacion_ingresos, bins=10, kde=False, color="#c9daa4", edgecolor="#62997a")

plt.axvline(media, color='#77779d', linestyle='--', linewidth=2, label=f'Media: {media:.2f}')
plt.axvline(mediana, color='#4771a3', linestyle='-', linewidth=2, label=f'Mediana: {mediana:.2f}')
plt.axvline(moda, color='#e7a8b1', linestyle=':', linewidth=2, label=f'Moda: {moda:.2f}')

plt.title("Distribución de ingresos simulados", fontsize=14)
plt.xlabel("Ingresos", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.xlim(4, 15)
plt.ylim(0, 8)
plt.show()
```

### Interpretación de resultados

*   La moda no es representativa en este ejercicio dado que la muestra aleatoria contiene datos continuos y no datos categóricos.

*   Como la media y la mediana tienen valores parecidos (10.21 y 10.53 respectivamente), se puede decir que la distribución de los datos es simétrica o casi simétrica.

### Cálculos de rango, varianza, desviación estándar y coeficiente de variación

```python
minimo = np.min(simulacion_ingresos)
maximo = np.max(simulacion_ingresos)

rango = maximo - minimo
varianza = np.var(simulacion_ingresos)
desviacion_estandar = np.std(simulacion_ingresos)
coeficiente_var = desviacion_estandar / media
```

### Visualización de resultados

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(simulacion_ingresos, bins=6, kde=True, color="#c9daa4", edgecolor="#62997a")

plt.axvline(media, color='#77779d', linestyle='--', linewidth=2, label=f'Media = {media:.2f}')
plt.axvline(media + desviacion_estandar, color='#d65981', linestyle=':', linewidth=1.5, label=f'+1σ = {media + desviacion_estandar:.2f}')
plt.axvline(media - desviacion_estandar, color='#d65981', linestyle=':', linewidth=1.5, label=f'-1σ = {media - desviacion_estandar:.2f}')
plt.axvline(minimo, color='#ff8c94', linestyle='-', linewidth=1.5, label=f'Mínimo = {minimo}')
plt.axvline(maximo, color='#ff8c94', linestyle='-', linewidth=1.5, label=f'Máximo = {maximo}')

plt.title("Distribución de ingresos")
plt.xlabel("Ingresos")
plt.ylabel("Frecuencia")
plt.xlim(3, 15)
plt.ylim(0, 9)
plt.legend()

texto = (
    f"Rango: {rango}\n"
    f"Varianza: {varianza:.2f}\n"
    f"Desviación estándar: {desviacion_estandar:.2f}\n"
    f"Coef. de variación: {coeficiente_var:.2f}"
)
plt.text(12.5, 6.5, texto, bbox=dict(facecolor='lightgreen', alpha=0.3), fontsize=10)

plt.tight_layout()
plt.show()
```

### Análisis de resultados

*   Los datos está distribuida entre [4.81, 14.50]

### Asimetría y curtosis

```python
asimetria = stats.skew(simulacion_ingresos)
curtosis  = stats.kurtosis(simulacion_ingresos)

print(f"Asimetría:  {asimetria:.4f}")
print(f"Curtosis:   {curtosis:.4f}")
```

### Visualización

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(simulacion_ingresos, bins=8, density=False, color="#c9daa4", edgecolor="#62997a",
        alpha=0.8, label='Muestra')

x = np.linspace(simulacion_ingresos.min() - 1, simulacion_ingresos.max() + 1, 200)
pdf = stats.norm.pdf(x, media, desviacion_estandar)
ax.plot(x, pdf * tamano_muestra * (simulacion_ingresos.max() - simulacion_ingresos.min()) / 8,
        color='green', linewidth=2, label='Distribución normal')

ax.axvline(media, color='steelblue', linestyle='--', linewidth=1.5,
           label=f'Media: {media:.2f}')

textstr = f'Asimetría: {asimetria:.2f}\nCurtosis:  {curtosis:.2f}'
ax.text(0.97, 0.75, textstr, transform=ax.transAxes,
        fontsize=11, verticalalignment='top', horizontalalignment='right',
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

ax.set_title('Distribución de la muestra vs. distribución normal')
ax.set_xlabel('Ingresos')
ax.set_ylabel('Frecuencia')
ax.legend()
plt.tight_layout()
plt.xlim(3, 16)
plt.ylim(0, 10)
plt.show()
```

### Análisis de resultados

Con respecto a la asimetría:

*   El valor es muy cercano a 0, lo que indica que la distribución es casi simétrica.
*   El signo negativo indica una ligera cola hacia la izquierda, lo que significa que hay algunos valores bajos que "jalan" ligeramente la distribución en esa dirección.
*   Al estar tan cerca de 0, se considera prácticamente simétrica, como una distribución normal.

Con respecto a curtosis:

*   El valor de -0.29 indica que la curva es técnicamente platicúrtica, pero visualmente el histograma muestra irregularidades, esto se puede deber al tamaño original de la muestra (30). Con más datos, la distribución tendería a suavizarse y acercarse más a la curva normal superpuesta.
*   Al ser tan cercano a 0, la diferencia con una normal es mínima.

### <span style="color:#2F749F;"><strong>Ejercicio 5: Análisis de correlación</strong></span>

En esta segunda parte, genera **dos variables** a partir de tú número de cédula y realiza un análisis **bivariado** para calcular la relación entre ellas.

1. Use los **6 últimos dígitos** de su cédula como base (similar al Ejercicio 1).  
2. Genere dos variables aleatorias (ejemplo: `ingresos` y `gastos`) con distribución normal, usando esa base como **semilla reproducible**.  
3. Calcule los coeficientes de correlación de **Pearson** y **Spearman**.
4. Grafique las variables analizadas (ver la imagen de ejemplo)
5. Interprete los resultados.  

image.png

```python
semilla = 415629
np.random.seed(semilla)

media_ingresos = (semilla % 10) + 2
media_gastos   = (semilla % 10) + 1

ingresos = np.random.normal(loc=media_ingresos, scale=2, size=30)
gastos   = np.random.normal(loc=media_gastos,   scale=1, size=30)
```

### Correlación de Pearson y Spearman

```python
pearson,  p_pears  = stats.pearsonr(ingresos, gastos)
spearman, p_spear  = stats.spearmanr(ingresos, gastos)

print(f"Coef. Pearson:  {pearson:.4f}  (p-valor: {p_pears:.4f})")
print(f"Coef. Spearman: {spearman:.4f}  (p-valor: {p_spear:.4f})")
```

### Visualización

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.scatter(ingresos, gastos, color="#c9daa4", edgecolor="#62997a",
           s=60, alpha=0.85)

m, b = np.polyfit(ingresos, gastos, 1)
x_line = np.linspace(ingresos.min(), ingresos.max(), 100)
ax.plot(x_line, m * x_line + b, color='red', linewidth=1.5,
        linestyle='--', alpha=0.7)

textstr = f'Coef. Pearson  = {pearson:.3f}\nCoef. Spearman = {spearman:.3f}'
ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=10,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='white', edgecolor='gray'))

ax.set_title('Relación entre ingresos y gastos', fontweight='bold')
ax.set_xlabel('Ingresos')
ax.set_ylabel('Gastos')
plt.tight_layout()
plt.show()
```

### Interpretación

*`Dirección y magnitud`*

Ambos coeficientes son positivos y similares ( aproximadamente 0.32), lo que indica una correlación positiva débil a moderada, es decir, cuando los ingresos tienden a ser mayores, los gastos también tienden a ser ligeramente mayores.

*`Significancia de los coeficientes`*

*   Pearson: p = 0.0828 > 0.05, no significativo
*   Spearman: p = 0.0801 > 0.05, no significativo

*`Comparación Pearson vs. Spearman`*

Los valores son casi idénticos (0.3219 y 0.3246), indicando que no hay valores atípicos influyentes ni una relación no lineal relevante.

*`Sobre el gráfico de dispersión`*

La línea de tendencia roja confirma la dirección positiva, pero la alta dispersión de los puntos alrededor de la línea explica la correlación débil. No se aprecia un patrón lineal claro.

### Conclusión

Existe una correlación positiva débil entre ingresos y gastos, pero no es estadísticamente significativa

### <span style="color:#2F749F;"><strong>Ejercicio 6: Chi-Cuadrado</strong></span>

En esta parte, analiza si existe **asociación estadística** entre dos variables categóricas simuladas a partir de tu número de cédula.

1. Usa los **6 últimos dígitos** de tu cédula como base para generar una semilla reproducible.
2. Simula una muestra de **60 personas** con dos variables categóricas:
   - `Nivel educativo`: Básico, Medio, Superior.
   - `Acceso a internet`: Sí, No.
3. Crea una **tabla de contingencia** entre ambas variables.
4. Aplica la **prueba de Chi cuadrado de independencia**.
5. Visualiza la distribución conjunta con un mapa de calor (ver la imagen de con la información de ejemplo).
6. Interpreta el resultado.

image.png

```python
semilla = 415629
np.random.seed(semilla)
```

### Simulación de 60 personas

```python
nivel_educativo = np.random.choice(['Básico', 'Medio', 'Superior'],
                                    size=60, p=[0.4, 0.35, 0.25])
acceso_internet = np.random.choice(['Sí', 'No'],
                                    size=60, p=[0.6, 0.4])

df = pd.DataFrame({'Nivel educativo': nivel_educativo,
                   'Acceso a internet': acceso_internet})
```

### Tabla de contingencia

```python
tabla = pd.crosstab(df['Nivel educativo'], df['Acceso a internet'])
print("Tabla de contingencia:")
print(tabla)
```

### Prueba de Chi2

```python
chi2, p_valor, gl, freq_esperadas = stats.chi2_contingency(tabla)
print(f"\nChi² = {chi2:.2f}, p-valor = {p_valor:.3f}, grados de libertad = {gl}")
print(f"\nFrecuencias esperadas:")
print(pd.DataFrame(freq_esperadas, index=tabla.index, columns=tabla.columns))
```

### Visualización con Heatmap

```python
ig, ax = plt.subplots(figsize=(7, 5))

sns.heatmap(tabla, annot=True, fmt='d', cmap='Greens',
            linewidths=0.5, ax=ax)

ax.set_title('Distribución conjunta: Nivel educativo vs. Acceso a internet',
             fontsize=11)
ax.set_xlabel('Acceso a internet')
ax.set_ylabel('Nivel educativo')
plt.tight_layout()
plt.show()
```

### Interpretación

Con un `Chi² = 2.60` y `p-valor = 0.273`, no se rechaza la hipótesis nula de independencia. Lo que no existe evidencia estadística suficiente para afirmar que hay asociación entre el nivel educativo y el acceso a internet en esta muestra simulada.

Aunque visualmente en el heatmap se aprecia una tendencia (verde más oscuro hacia "Sí" en niveles `Medio` y `Superior`), las diferencias no son estadísticamente significativas. Esto puede deberse a:

*   El tamaño muestral pequeño (*n=60*) limita la potencia estadística.
*   Las frecuencias esperadas (`9.1–11.9`) son similares entre grupos, lo que reduce el estadístico Chi².
*   Los datos son simulados aleatoriamente, por lo que no necesariamente reflejan un patrón real.

El dataset con el que trabajaremos será con palmerpenguins en su versión simplificada.

A continuación una muestra de los datos:

```python
penguins = sns.load_dataset("penguins")
penguins = penguins.dropna()
penguins.head()
```

<h3 style="background-color:#e6d1ca; padding:10px; border-radius:5px; display:inline-block; width:auto;">
<b>Análisis de PCA</b>
</h3>

Realizaremos este análisis buscando poder:


*   Identificar si hay correlaciones fuertes entre medidas morfológicas como el largo del pico, profundidad del pico, longitud de la aleta y masa corporal.
*   Transformar las 4 variables numéricas originales (b*ill_length_mm*, *bill_depth_mm*, *flipper_length_mm*, *body_mass_g*) en componentes principales que capturen la mayor parte de la variabilidad.

*   Visualizar agrupamientos entre especies (Adelie, Chinstrap, Gentoo) en un espacio reducido de dos dimensiones
*   Interpretar qué variables morfológicas contribuyen más a cada componente principal, usando species como variable de referencia para colorear y validar los agrupamientos









```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

numeric_cols = penguins.select_dtypes(include=[np.number]).columns

#Estandarizamos las variables
scaler = StandardScaler()
X_std = scaler.fit_transform(penguins[numeric_cols])

pca = PCA()
X_pca = pca.fit_transform(X_std)

pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(len(numeric_cols))])
pca_df['species'] = penguins['species'].values

sns.pairplot(penguins, hue='species', palette='Set2', diag_kind='hist', corner=False)
plt.suptitle('Matriz de dispersión: relaciones entre variables de Penguins', fontsize=16, y=1.02)
plt.show()
```

Según la matriz de dispersión, podemos concluir que la especie `Gentoo` tiende a diferenciarse bastante bien respecto a las otras 2 especies. Además de esto, podemos observar que la especie `Gentoo` es más grande dado que tanto su masa como el tamaño de su aleta, son más grandes en comparación con las otras 2 especies.

También podemos observar que hay una correlación bastante notoria entre el peso del pinguino y el tamaño de su aleta. Esto aplica para las 3 especies, y tiene sentido dado que pinguinos más pesados tienen aletas más largas, lo que explica una proporcionalidad en los cuerpos.

```python
# Matriz de componentes
componentes = pd.DataFrame(
    pca.components_.T,
    columns=[f'PC{i+1}' for i in range(len(numeric_cols))],
    index=numeric_cols
)

# Visualización de autovectores
plt.figure(figsize=(8,5))
sns.heatmap(componentes, annot=True, cmap='coolwarm', center=0)
plt.title('Autovectores - Penguins')
plt.ylabel('Variables originales')
plt.xlabel('Componentes principales')
plt.tight_layout()
plt.show()

# Ecuaciones de los componentes principales con nombres reales
print("Ecuaciones de los componentes principales:\n")

for i, comp in enumerate(componentes.columns):
    pesos = pca.components_[i]
    ecuacion = " + ".join(
        [f"{round(pesos[j],3)}·'{numeric_cols[j]}'" for j in range(len(pesos))]
    )
    print(f"{comp} = {ecuacion}")
```

Según nuestro mapa de calor, podemos resaltar varias anotaciones.


*   El PC1, representa el tamaño general del pinguino, pues los pesos de su masa, el largo de su aleta y el largo de su pico, son relativamente altos
*   El PC2 se concentra en la forma del pico, pues los pesos del largo del pico y de su profundidad, son relativamente altos. En especial el último mencionado.

Hasta este punto, podríamos decir que estos 2 componentes principales son bastante buenos, pues están describiendo en su totalidad todas nuestras características. Sin embargo, sigamos con el análisis de los de los otros componentes principales:

*  PC3 representa el largo del pico exclusivamente y muestra una relación inversa entre esta y la masa del pinguino

*  PC4 representa el largo de la aleta exclusivamente, y muestra una relación inversa entre esta y la masa del pinguino



```python
# Varianza explicada
varianza = pca.explained_variance_
PVE = pca.explained_variance_ratio_
PVE_acum = np.cumsum(PVE)
componentes = [f'PC{i+1}' for i in range(len(PVE))]

# DataFrame para visualización
df_varianza = pd.DataFrame({
    'Componente': componentes,
    'Autovalor': varianza,
    'PVE': PVE,
    'PVE acumulada': PVE_acum
})

print("Matriz resumen de varianza explicada:\n")
print(df_varianza.round(3))


# Gráfica de autovalores
plt.figure(figsize=(8,5))
ax = sns.barplot(x='Componente', y='Autovalor', data=df_varianza, color='lightsteelblue')

for i, v in enumerate(df_varianza['Autovalor']):
    ax.text(i, v + 0.05, f"{v:.2f}", ha='center', fontsize=9)

plt.title("Autovalores - Penguins", fontsize=15, fontweight="bold",
          backgroundcolor="#fff9db", pad=15)
plt.ylabel("Autovalor (λ)")
plt.xlabel("Componentes principales")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()


# Gráfica de varianza explicada
plt.figure(figsize=(8,5))

sns.barplot(x='Componente', y='PVE', data=df_varianza,
            color='skyblue', label='PVE individual')

sns.lineplot(x='Componente', y='PVE acumulada',
             data=df_varianza, marker='o',
             color='darkblue', label='PVE acumulada')

plt.axhline(0.7, color='red', linestyle='--', label='Umbral 70%')
plt.axhline(0.9, color='green', linestyle='--', label='Umbral 90%')

for i, v in enumerate(df_varianza['PVE']):
    plt.text(i, v + 0.01, f"{v:.2f}", ha='center', fontsize=9)

plt.title("Proporción de varianza explicada - Penguins",
          fontsize=15, fontweight="bold",
          backgroundcolor="#fff9db", pad=15)

plt.ylabel("Proporción de varianza")
plt.xlabel("Componentes principales")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

<b>Análisis con Regresión Lineal Múltiple</b>
</h3>

En este ejemplo se realizó una regresión lineal múltiple usando mínimos cuadrados ordinarios (OLS), con el objetivo de:



*   Estimar la relación entre las variables morfológicas (*bill_length_mm*, *bill_depth_mm,* *flipper_length_mm)* y la masa corporal (*body_mass_g*)
*   Interpretar los coeficientes del modelo para identificar qué medida tiene mayor efecto sobre la masa corporal
*   Evaluar la bondad de ajuste mediante el R² y la significancia estadística de cada variable (valor p)
*   Analizar los residuos para verificar si el modelo cumple los supuestos de linealidad y homocedasticidad










```python
# Variables predictoras
X = penguins[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm']]

# Variable respuesta
Y = penguins['body_mass_g']

# Agregar constante para el intercepto
X_const = sm.add_constant(X)

# Ajustar modelo de regresión lineal múltiple
model = sm.OLS(Y, X_const).fit()

# Mostrar resumen del modelo
print(model.summary())
```

Los resultados de los mínimos cuadrados ordinarios nos indican que:

*  El *R² *de 0.764 indica que el 76.4% de la varianza en la masa corporal es explicada por las tres variables morfológicas, lo que representa un buen ajuste del modelo.
*  El *F-statistic* de 354.9 con p = 9.26e-103 confirma que el modelo en conjunto es estadísticamente significativo.


*   El largo y profundidad del pico no son estadísticamente significativas (p = 0.540 y p = 0.198, ambos > 0.05), por lo que no aportan de forma confiable al modelo.
*   El *Cond. No* de 5440 es alto, lo que sugiere que puede haber multicolinealidad entre las variables predictoras, es decir, que algunas están correlacionadas entre sí, lo que dificulta estimar los coeficientes de forma precisa.







```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Gráfica 1: Valores reales vs predichos
sns.scatterplot(x=Y, y=model.fittedvalues, color='#9b5f7b', ax=axes[0])
axes[0].plot([Y.min(), Y.max()], [Y.min(), Y.max()], 'r--')
axes[0].set_xlabel('Masa corporal real (g)')
axes[0].set_ylabel('Masa corporal predicha (g)')
axes[0].set_title('Gráfico 1: Valores reales vs predichos')

# Gráfica 2: Residuos vs valores ajustados
sns.scatterplot(x=model.fittedvalues, y=model.resid, color='#525574', ax=axes[1])
axes[1].axhline(0, color='red', linestyle='--')
axes[1].set_xlabel('Valores ajustados')
axes[1].set_ylabel('Residuos')
axes[1].set_title('Gráfico 2: Residuos vs valores ajustados')

plt.tight_layout()
plt.show()
```

### Sobre el gráfico #1: Valores reales vs predichos

*   Los puntos siguen de cerca la línea roja diagonal, lo que confirma que el modelo tiene un buen ajuste general, consistente con el R² de 0.764 del punto anterior.

### Sobre el gráfico 2: Residuos vs valores ajustados

*   Los residuos se distribuyen de forma relativamente aleatoria alrededor de cero, lo que indica que el modelo cumple el supuesto de homocedasticidad (varianza constante de los errores). (2)

*   No se observa un patrón claro en forma de curva o embudo, indicando que el modelo es válido


  *   Una distribución en forma de curva indicaría que la relación entre variables no es lineal, es decir, que el mode de RLM no sería ideal para estos datos.
  *   Una distribución en forma de embudo indicaría que la varianza de los errores no es constante, violando uno de los supuestos de la regresión lineal. (2)
