A partir de una muestra simulada de ingresos, calcula la **media, mediana y moda**, y luego a interpretarlas.

1. Usa el número base `528310` (debes reemplazarlo con los últimos 6 dígitos de tu documento de identificación).  
2. Con este número genera una semilla (`np.random.seed`) para que los resultados sean reproducibles.  
3. Los ingresos se simulan con una distribución normal:

   - Desviación estándar: `2`  
   - Tamaño de muestra: `30`

### Importación de librerías

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
import pandas as pd
```

### Generación de ingresos simulados con semilla: 911848

```python
semilla = 911848
np.random.seed(semilla)

media_teorica = (semilla % 10) + 2
desviacion_estandar = 2
tamano_muestra = 30

simulacion_ingresos = np.random.normal(loc=media_teorica, scale=desviacion_estandar, size=tamano_muestra)

print(simulacion_ingresos)
```

### Cálculo de medidas de tendencia central

```python
media = np.mean(simulacion_ingresos)
mediana = np.median(simulacion_ingresos)
moda = stats.mode(simulacion_ingresos, keepdims=True)[0][0]
```

### Gráfica de resultados

```python
plt.figure(figsize=(10, 5))
sns.histplot(simulacion_ingresos, bins=10, kde=False, color="#c9daa4", edgecolor="#62997a")

plt.axvline(media, color='#77779d', linestyle='--', linewidth=2, label=f'Media: {media:.2f}')
plt.axvline(mediana, color='#4771a3', linestyle='-', linewidth=2, label=f'Mediana: {mediana:.2f}')
plt.axvline(moda, color='#e7a8b1', linestyle=':', linewidth=2, label=f'Moda: {moda:.2f}')

plt.title("Distribución de ingresos simulados", fontsize=14)
plt.xlabel("Ingresos", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.xlim(4, 15)
plt.ylim(0, 8)
plt.show()
```

### Interpretación de resultados

*   La moda no es representativa en este ejercicio dado que la muestra aleatoria contiene datos continuos y no datos categóricos.

*   Como la media y la mediana tienen valores parecidos (10.21 y 10.53 respectivamente), se puede decir que la distribución de los datos es simétrica o casi simétrica.

### Cálculos de rango, varianza, desviación estándar y coeficiente de variación

```python
minimo = np.min(simulacion_ingresos)
maximo = np.max(simulacion_ingresos)

rango = maximo - minimo
varianza = np.var(simulacion_ingresos)
desviacion_estandar = np.std(simulacion_ingresos)
coeficiente_var = desviacion_estandar / media
```

### Visualización de resultados

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(simulacion_ingresos, bins=6, kde=True, color="#c9daa4", edgecolor="#62997a")

plt.axvline(media, color='#77779d', linestyle='--', linewidth=2, label=f'Media = {media:.2f}')
plt.axvline(media + desviacion_estandar, color='#d65981', linestyle=':', linewidth=1.5, label=f'+1σ = {media + desviacion_estandar:.2f}')
plt.axvline(media - desviacion_estandar, color='#d65981', linestyle=':', linewidth=1.5, label=f'-1σ = {media - desviacion_estandar:.2f}')
plt.axvline(minimo, color='#ff8c94', linestyle='-', linewidth=1.5, label=f'Mínimo = {minimo}')
plt.axvline(maximo, color='#ff8c94', linestyle='-', linewidth=1.5, label=f'Máximo = {maximo}')

plt.title("Distribución de ingresos")
plt.xlabel("Ingresos")
plt.ylabel("Frecuencia")
plt.xlim(3, 15)
plt.ylim(0, 9)
plt.legend()

texto = (
    f"Rango: {rango}\n"
    f"Varianza: {varianza:.2f}\n"
    f"Desviación estándar: {desviacion_estandar:.2f}\n"
    f"Coef. de variación: {coeficiente_var:.2f}"
)
plt.text(12.5, 6.5, texto, bbox=dict(facecolor='lightgreen', alpha=0.3), fontsize=10)

plt.tight_layout()
plt.show()
```

### Análisis de resultados

*   Los datos está distribuida entre [4.81, 14.50]
*   Con el rango, podemos decir que los datos están concentrados alrededor de 10.
*   El coeficiente de variación de 0.21 (21%) indica que la dispersión es baja-moderada, lo que significa que los datos son relativamente homogéneos respecto a la media.
*   La distribución visual luce aproximadamente simétrica y cercana a una normal, lo cual es esperado por el uso de *np.random.normal* en el ejercicio 1.








### <span style="color:#2F749F;"><strong>Ejercicio 3: Cálculo de medidas de distribución</strong></span>

### Asimetría y curtosis

```python
asimetria = stats.skew(simulacion_ingresos)
curtosis  = stats.kurtosis(simulacion_ingresos)

print(f"Asimetría:  {asimetria:.4f}")
print(f"Curtosis:   {curtosis:.4f}")
```

### Visualización

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(simulacion_ingresos, bins=8, density=False, color="#c9daa4", edgecolor="#62997a",
        alpha=0.8, label='Muestra')

x = np.linspace(simulacion_ingresos.min() - 1, simulacion_ingresos.max() + 1, 200)
pdf = stats.norm.pdf(x, media, desviacion_estandar)
ax.plot(x, pdf * tamano_muestra * (simulacion_ingresos.max() - simulacion_ingresos.min()) / 8,
        color='green', linewidth=2, label='Distribución normal')

ax.axvline(media, color='steelblue', linestyle='--', linewidth=1.5,
           label=f'Media: {media:.2f}')

textstr = f'Asimetría: {asimetria:.2f}\nCurtosis:  {curtosis:.2f}'
ax.text(0.97, 0.75, textstr, transform=ax.transAxes,
        fontsize=11, verticalalignment='top', horizontalalignment='right',
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

ax.set_title('Distribución de la muestra vs. distribución normal')
ax.set_xlabel('Ingresos')
ax.set_ylabel('Frecuencia')
ax.legend()
plt.tight_layout()
plt.xlim(3, 16)
plt.ylim(0, 10)
plt.show()
```

### Análisis de resultados

Con respecto a la asimetría:

*   El valor es muy cercano a 0, lo que indica que la distribución es casi simétrica.
*   El signo negativo indica una ligera cola hacia la izquierda, lo que significa que hay algunos valores bajos que "jalan" ligeramente la distribución en esa dirección.
*   Al estar tan cerca de 0, se considera prácticamente simétrica, como una distribución normal.

Con respecto a curtosis:

*   El valor de -0.29 indica que la curva es técnicamente platicúrtica, pero visualmente el histograma muestra irregularidades, esto se puede deber al tamaño original de la muestra (30). Con más datos, la distribución tendería a suavizarse y acercarse más a la curva normal superpuesta.
*   Al ser tan cercano a 0, la diferencia con una normal es mínima.

### <span style="color:#2F749F;"><strong>Ejercicio 5: Análisis de correlación</strong></span>

En esta segunda parte, genera **dos variables** a partir de tú número de cédula y realiza un análisis **bivariado** para calcular la relación entre ellas.

1. Use los **6 últimos dígitos** de su cédula como base (similar al Ejercicio 1).  
2. Genere dos variables aleatorias (ejemplo: `ingresos` y `gastos`) con distribución normal, usando esa base como **semilla reproducible**.  
3. Calcule los coeficientes de correlación de **Pearson** y **Spearman**.
4. Grafique las variables analizadas (ver la imagen de ejemplo)
5. Interprete los resultados.  

image.png

```python
semilla = 416823
np.random.seed(semilla)

media_ingresos = (semilla % 10) + 2
media_gastos   = (semilla % 10) + 1

ingresos = np.random.normal(loc=media_ingresos, scale=2, size=30)
gastos   = np.random.normal(loc=media_gastos,   scale=1, size=30)
```

### Correlación de Pearson y Spearman

```python
pearson,  p_pears  = stats.pearsonr(ingresos, gastos)
spearman, p_spear  = stats.spearmanr(ingresos, gastos)

print(f"Coef. Pearson:  {pearson:.4f}  (p-valor: {p_pears:.4f})")
print(f"Coef. Spearman: {spearman:.4f}  (p-valor: {p_spear:.4f})")
```

### Visualización

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.scatter(ingresos, gastos, color="#c9daa4", edgecolor="#62997a",
           s=60, alpha=0.85)

m, b = np.polyfit(ingresos, gastos, 1)
x_line = np.linspace(ingresos.min(), ingresos.max(), 100)
ax.plot(x_line, m * x_line + b, color='red', linewidth=1.5,
        linestyle='--', alpha=0.7)

textstr = f'Coef. Pearson  = {pearson:.3f}\nCoef. Spearman = {spearman:.3f}'
ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=10,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='white', edgecolor='gray'))

ax.set_title('Relación entre ingresos y gastos', fontweight='bold')
ax.set_xlabel('Ingresos')
ax.set_ylabel('Gastos')
plt.tight_layout()
plt.show()
```

### Conclusión

No existe relación entre ingresos y gastos en la muestra simulada. Dado que laas variables se generan independientemente, los coeficientes serán cercanos a 0, indicando poca o ninguna correlación lineal.

### <span style="color:#2F749F;"><strong>Ejercicio 6: Chi-Cuadrado</strong></span>

En esta parte, analiza si existe **asociación estadística** entre dos variables categóricas simuladas a partir de tu número de cédula.

1. Usa los **6 últimos dígitos** de tu cédula como base para generar una semilla reproducible.
2. Simula una muestra de **60 personas** con dos variables categóricas:
   - `Nivel educativo`: Básico, Medio, Superior.
   - `Acceso a internet`: Sí, No.
3. Crea una **tabla de contingencia** entre ambas variables.
4. Aplica la **prueba de Chi cuadrado de independencia**.
5. Visualiza la distribución conjunta con un mapa de calor (ver la imagen de con la información de ejemplo).
6. Interpreta el resultado.

image.png

```python
semilla = 416823
np.random.seed(semilla)
```

### Simulación de 60 personas

```python
nivel_educativo = np.random.choice(['Básico', 'Medio', 'Superior'],
                                    size=60, p=[0.4, 0.35, 0.25])
acceso_internet = np.random.choice(['Sí', 'No'],
                                    size=60, p=[0.6, 0.4])

df = pd.DataFrame({'Nivel educativo': nivel_educativo,
                   'Acceso a internet': acceso_internet})
```

### Tabla de contingencia

```python
tabla = pd.crosstab(df['Nivel educativo'], df['Acceso a internet'])
print("Tabla de contingencia:")
print(tabla)
```

### Prueba de Chi2

```python
chi2, p_valor, gl, freq_esperadas = stats.chi2_contingency(tabla)
print(f"\nChi² = {chi2:.2f}, p-valor = {p_valor:.3f}, grados de libertad = {gl}")
print(f"\nFrecuencias esperadas:")
print(pd.DataFrame(freq_esperadas, index=tabla.index, columns=tabla.columns))
```

### Visualización con Heatmap

```python
ig, ax = plt.subplots(figsize=(7, 5))

sns.heatmap(tabla, annot=True, fmt='d', cmap='Greens',
            linewidths=0.5, ax=ax)

ax.set_title('Distribución conjunta: Nivel educativo vs. Acceso a internet',
             fontsize=11)
ax.set_xlabel('Acceso a internet')
ax.set_ylabel('Nivel educativo')
plt.tight_layout()
plt.show()
```

### Conclusión

Con un Chi² = 0.18 y p-valor = 0.915, no se rechaza la hipótesis nula de independencia. Lo que significa que el nivel educativo y el acceso a internet son variables estadísticamente independientes en esta muestra, es decir, conocer el nivel educativo de una persona no permite predecir si tiene o no acceso a internet.

El resultado es esperado dado que ambas variables fueron simuladas de forma aleatoria e independiente con *np.random.choice*
