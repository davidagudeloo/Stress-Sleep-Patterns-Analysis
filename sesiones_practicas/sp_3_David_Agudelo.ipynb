**Carga de datos y conversión a dataframe**

```python
from sklearn.datasets import fetch_california_housing
casas_california = fetch_california_housing()
df_casas_california = pd.DataFrame(casas_california.data, columns= casas_california.feature_names)
df_casas_california["target"] = casas_california.target
df_casas_california
```

### 3



**Variables seleccionadas:**

   - `MedInc`: ingreso medio de la zona
   - `HouseAge`: antigüedad promedio de la casa
  - `AveRooms`: promedio de habitaciones

### 4
**Normalización**


**Normalización L1**

```python
from sklearn.preprocessing import Normalizer
columnas = ['MedInc', 'HouseAge', 'AveRooms']
# Normalización L1
normalizador_l1 = Normalizer(norm='l1')
l1_mis_variables = normalizador_l1.fit_transform(df_casas_california[columnas])
```

**Normalización L2**

```python
# Normalización L2
normalizador_l2 = Normalizer(norm='l2')
l2_mis_variables = normalizador_l2.fit_transform(df_casas_california[columnas])
```

### 5



```python
# import numpy as np
# import matplotlib.pyplot as plt


# Preparar datos para graficar
indices = np.arange(1, len(df_casas_california[['MedInc']]) + 1)
valores_originales = df_casas_california[['MedInc']].values.flatten()
valores_escalados_l1 = l1_mis_variables[:,0]
valores_escalados_l2 = l2_mis_variables[:,0]

# Obtener min y max originales
x_min = np.min(valores_originales)
x_max = np.max(valores_originales)

# Crear figura y subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Gráfico 1: Datos originales en toda la columna 1
axs[0,0].scatter(indices, valores_originales, color='steelblue')
axs[0,0].set_title('Distribución original')
axs[0,0].set_xlabel('Índice')
axs[0,0].set_ylabel('Valor')
axs[0,0].grid(True)

axs[1,0].scatter(indices, valores_originales, color='steelblue')
axs[1,0].set_title('Distribución original')
axs[1,0].set_xlabel('Índice')
axs[1,0].set_ylabel('Valor')
axs[1,0].grid(True)








# Gráfico 2: Datos escalonados
axs[0,1].scatter(indices, valores_escalados_l1, color='red')
axs[0,1].set_title('Datos escalados L1')
axs[0,1].set_xlabel('Índice')
axs[0,1].set_ylabel('Valor Escalado')
axs[0,1].grid(True)

# Gráfico 3: Datos escalonados
axs[1,1].scatter(indices, valores_escalados_l2, color='green')
axs[1,1].set_title('Datos escalados L2')
axs[1,1].set_xlabel('Índice')
axs[1,1].set_ylabel('Valor Escalado')
axs[1,1].grid(True)




# Anotaciones de min y max
valores_lista = [valores_escalados_l1,valores_escalados_l2]
for i in range(2):
  axs[i,1].text(1, 0.9, f' Máx= {np.max(valores_lista[i]):.3f}', transform=axs[i,1].transAxes, fontsize=10, color='darkred', va='top')
  axs[i,1].text(1, 1.0, f' Mín= {np.min(valores_lista[i]):.3f}', transform=axs[i,1].transAxes, fontsize=10, color='darkgreen', va='top')
  axs[i,0].text(1, 0.9, f' Máx= {x_max:.2f}', transform=axs[i,0].transAxes, fontsize=10, color='darkred', va='top')
  axs[i,0].text(1, 1.0, f' Mín= {x_min:.2f}', transform=axs[i,0].transAxes, fontsize=10, color='darkgreen', va='top')

# Mostrar los gráficos
plt.tight_layout()
plt.show()
```

### 6


Para este ejemplo, contrastando la normalización para el ingreso promedio por zona, observamos que  la normalización L1 genera una distribución más comprimida respecto a la normalización L2. Dependiendo de qué tan grandes o pequeños son los valores entre las features, L1 podría tender a llevar muchos de los pesos de algunas variables a valores muy cercanos a cero, mientras que L2, prefiere mantener todos los pesos pequeños, pero no exactamente en cero. Sin conocer a fondo la naturaleza de los otros datos, recomendaría L2, ya que conserva mejor la estructura original de los datos.

### 6


La transformación logarítmica comprime los valores altos, lo cual puede resultar útil cuando hay valores extremos, mientras que la logarítmica reflejada y la cuadrática se usan cuando la distribución tiene cola a la izquierda (asimetría negativa). La transformación cuadrática amplifica las diferencias, por lo que se ve un aumento en la dispersión en la gráfica. Por otro lado, Box-Cox ajusta automáticamente la forma de la distribución para hacerla más cercana a una normal. Recomendaría Box-Cox porque logra una distribución más equilibrada y adecuada para análisis estadístico y para procesos futuros de ingreso de data a modelos.
